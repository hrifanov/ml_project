```python
import numpy as np
import pandas as pd 

import os
import glob

import cv2
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import img_to_array
from keras.utils import np_utils
from numba import jit, cuda
```

```python
coordinates_path = '../../Dataset/list_bbox_celeba.txt'
photos_path = '../../Dataset/img_celeba/'
```

```python
image_names = os.listdir(photos_path)
image_names.sort()
data   = pd.DataFrame(image_names, columns=['file_name'])

labels = pd.read_csv(coordinates_path, sep='\s+')
labels.drop('image_id', axis=1, inplace=True)
```

```python
print(data.shape)
print(labels.shape)
```

```python
def load_images(photos_path, LIMIT):
    data = []
    paths = glob.glob(f"{photos_path}*.jpg")[:LIMIT]
    
    images = [cv2.imread(file) for file in paths]
    
    for img in images:
        try:
            img_array = img_to_array(img)
            data.append(img_array)
        except:
            None
            
    return data
```

```python
train_size = 0.6


x_train, x_rem, y_train, y_rem = train_test_split(data, labels, train_size=train_size, random_state = 101, shuffle=True)

x_valid, x_test, y_valid, y_test = train_test_split(x_rem,y_rem, test_size=0.5)
```

```python
print(x_train.shape)
print(x_valid.shape)
print(x_test.shape)
```

```python
file_name = x_train.iloc[2]['file_name']
img = cv2.imread(photos_path + file_name)

label = y_train.iloc[2]

x1 = label['x_1']
y1 = label['y_1']
x2 = x1 + label['width']
y2 = y1 + label['height']

UPPER_LEFT = (x1, y1)
BOTTOM_RIGHT = (x2, y2)

print(file_name)
```

```python
BOUNDING_BOX_COLOR = (0, 0, 255)
BOUNDING_BOX_THICKNESS = 2

rectangle = cv2.rectangle(img.copy(), UPPER_LEFT, BOTTOM_RIGHT, (0, 0, 255), BOUNDING_BOX_THICKNESS)
plt.imshow(cv2.cvtColor(rectangle, cv2.COLOR_BGR2RGB))
```
